Add new column (Days)

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Duration of stay (Days)'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
df['Days'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

```python
plt.figure(figsize=(17,8))
n, bins, patches = plt.hist(x=df['Duration of stay (Days)'], bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)
cmap = plt.get_cmap('Blues')
for c, p in zip(bins, patches):
    plt.setp(p, 'facecolor', cmap(c))
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Duration of stay (Days)')
plt.ylabel('Number of Patients')
plt.title('Distribution of Duration of stay (Days)')
plt.show()
```

```python
plt.figure(figsize=(17,8))
cmap = plt.get_cmap('Blues')
n, bins, patches = plt.hist(x=df['Days'], bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)
norm = plt.Normalize(df['Days'].min(), df['Days'].max())
for c, p in zip(bins, patches):
    plt.setp(p, 'facecolor', cmap(norm(c)))
    plt.grid(axis='y', alpha=0.75)
plt.xlabel('Days')
plt.ylabel('Number of Patients')
plt.title('Distribution of Days')
plt.show()
```
